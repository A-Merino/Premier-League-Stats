### Background/Why
In football (soccer), players get transferred to different teams all the time, each team averages around 10 total transfers (5 players in and 5 players out). Unlike other sports and leagues such as the NBA, MLB, etc. where they mostly do player trades or wait until free agency to move to a different team in combination with a big contract signing with it. In European football trading players almost never occurs, but signing players that don't have a team does quite often. But when a team wants to buy a player from another team, they want to pay a fair price for that player without paying more than that player is worth. The team selling, however, has the same dilemma except that they want to sell the player for more than they are worth to get a few extra bucks to reinvest into other players, facilities, etc.. Those values that the teams base their starting prices off are more times than not the **transfer value** of a player. A transfer value is a price that is obtained after combining a bunch of data and a little but of public opinion (Brazilian players are much more highly regarded than Polish players). 

According to the [website](https://www.transfermarkt.co.in/transfermarkt-market-value-explained-how-is-it-determined-/view/news/385100) I pulled all of my values from, some of the most important factors to a transfer value are:
* Future prospects
* Age
* Performance at the club and national team
* Level and status of the league, both in sporting and financial terms
* Reputation/prestige
* Development potential
* League-specific features
* Marketing value
* Number & reputation of interested clubs
* Performance potential
* Experience level
* Injury susceptibility
* Different financial conditions of clubs and leagues
* General demand and "trends" on the market
* General development of transfer fees
* External factors such as the coronavirus pandemic and its consequences

Many of these are based off opinion and are somewhat intangible such as "Injury Susceptibility", but I wanted to focus on just the performances of players throughout a season. The value of a player is always changing based on their recent run of performances, but to get a value that accurately represents how a player did in a season, I will be using the value from the summer after the season ends (May). 

The data was obtained from a website called FBref, which contained easy to scrape tables full of many datapoints for each team in the Premier League, which is the league of focus. An example of the page of one team is [here](https://fbref.com/en/squads/18bb7c10/2019-2020/Arsenal-Stats). Statistics has begun to show up a lot more recently in football (similar to Billy Bean and Moneyball), so there have been new metrics that have been created such as expected goals. The earliest use of that metric and others began in the 2017/18 season. From that year onwards is going to be the dataset for this problem, which is 6 years worth of football statistics such as:
* Age
* Position
* Goals
* Minutes Played
* Passes (Number of short, medium, long)
* and just under 150 more

The goal of this is to see if the value of these players still hold up when the public opinion and intangibles are stripped away, of if the data tells a very different story about the prices of certain players.

# Total Dataset

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error 
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor, plot_tree
from xgboost import XGBRegressor
```

```python
# Create inputs and outputs and split them
X = data.drop(["Player","Value"], axis=1).astype('float32')
Y = data.iloc[:,-1].astype('float32')
trainX, testX, trainY, testY = train_test_split(X, Y, test_size=0.25, random_state=38)
```

```python
# Run a Linear Regression Model
linModel1 = LinearRegression()
linModel1.fit(trainX, trainY)

predY = linModel1.predict(testX)

print("The Mean Absolute Error is:", mean_absolute_error(predY, testY))
print("The Mean Squared Error is:", mean_squared_error(predY, testY))
outs = range(len(predY))
plt.plot(outs, predY, "ro", label="Predicted")
plt.plot(outs, testY, "bo", label="True")
plt.legend()
plt.show()
```

```python
# Run a Ridge Regression Model
rdgModel1 = Ridge(alpha=0.1, solver="lsqr")
rdgModel1.fit(trainX, trainY)

predY = rdgModel1.predict(testX)

print("The Mean Absolute Error is:", mean_absolute_error(predY, testY))
print("The Mean Squared Error is:", mean_squared_error(predY, testY))
outs = range(len(predY))
plt.plot(outs, predY, "r", label="Predicted")
plt.plot(outs, testY, "b", label="True")
plt.legend()
plt.show()
```

```python
# Run a K-Nearest Neighbors model
knnModel1 = KNeighborsRegressor(n_neighbors=5, n_jobs=-1)
knnModel1.fit(trainX, trainY)

predY = knnModel1.predict(testX)

print("The Mean Absolute Error is:", mean_absolute_error(predY, testY))
print("The Mean Squared Error is:", mean_squared_error(predY, testY))
outs = range(len(predY))
plt.plot(outs, predY, "r", label="Predicted")
plt.plot(outs, testY, "b", label="True")
plt.legend()
plt.show()
```

```python
# Run a decision tree model
dtModel1 = DecisionTreeRegressor(random_state=38)
dtModel1.fit(trainX, trainY)

predY = dtModel1.predict(testX)

print("The Mean Absolute Error is:", mean_absolute_error(predY, testY))
print("The Mean Squared Error is:", mean_squared_error(predY, testY))
outs = range(len(predY))
plt.plot(outs, predY, "r", label="Predicted")
plt.plot(outs, testY, "b", label="True")
plt.legend()
plt.show()
```

```python
# Plot the tree it makes
plt.figure(figsize=(20,12))
plot_tree(dtModel1, filled=True, fontsize=10)
plt.show()
```

```python
# Run a random Forest model
rfModel1 = RandomForestRegressor(n_estimators=10, n_jobs=-1, random_state=38)
rfModel1.fit(trainX, trainY)

predY = rfModel1.predict(testX)

print("The Mean Absolute Error is:", mean_absolute_error(predY, testY))
print("The Mean Squared Error is:", mean_squared_error(predY, testY))
outs = range(len(predY))
plt.plot(outs, predY, "r", label="Predicted")
plt.plot(outs, testY, "b", label="True")
plt.legend()
plt.show()
```

```python
# Run an Ada Boost model
adaModel1 = AdaBoostRegressor(n_estimators=25, random_state=38)
adaModel1.fit(trainX, trainY)

predY = adaModel1.predict(testX)

print("The Mean Absolute Error is:", mean_absolute_error(predY, testY))
print("The Mean Squared Error is:", mean_squared_error(predY, testY))
outs = range(len(predY))
plt.plot(outs, predY, "r", label="Predicted")
plt.plot(outs, testY, "b", label="True")
plt.legend()
plt.show()
```

```python
# Run an XGBoost Model
xgbModel1 = XGBRegressor(gamma=1, subsample=0.5, reg_lambda=1, alpha=0.5)
xgbModel1.fit(trainX, trainY)

predY = xgbModel1.predict(testX)

print("The Mean Absolute Error is:", mean_absolute_error(predY, testY))
print("The Mean Squared Error is:", mean_squared_error(predY, testY))
outs = range(len(predY))
plt.plot(outs, predY, "ro", label="Predicted")
plt.plot(outs, testY, "bo", label="True")
plt.legend()
plt.show()
```

```python
close = 0
# Collect the number of players that were incorrectly predected by over 25
total_wrong = []
tester = dict(testY)
for i, part in enumerate(tester):
    if abs(predY[i]-tester[part]) > 25:
        total_wrong.append((part,predY[i]))
    if abs(predY[i]-tester[part]) < 1.8:
        close += 1 # Printing how many are within a certain range
close
```

# By Position

One of the most important skills in soccer is to play your position and to play it extremely well. Each position has its own very important job, which is why defenders don't score goals often and forwards are not since throwing their bodies in front of the ball when the opposing team shoots on their goal. The goal by spliting the dataset into four smaller ones is to see if a value can be more accurately predicted when comparing players to other players that only play the same position.

The four positions are:
* Goalkeeper
* Defender
* Midfielder
* Forward

Each position within each category (excluding goalkeeper) does have many different subcategories within, but overall they have relatively the same roles.

```python
# Split into four subsets based on position
forwards = pd.read_csv("player_data/fw_dataset.csv", encoding="latin")
mids = pd.read_csv("player_data/mf_dataset.csv", encoding="latin")
defends = pd.read_csv("player_data/df_dataset.csv", encoding="latin")
gks = pd.read_csv("player_data/gk_dataset.csv", encoding="latin")

# Drop columns and create inputs and outputs
fX = forwards.drop(["Player", "Pos","Value"], axis=1)
fY = forwards.iloc[:,-1]
mX = mids.drop(["Player", "Pos","Value"], axis=1)
mY = mids.iloc[:,-1]
dX = defends.drop(["Player", "Pos","Value"], axis=1)
dY = defends.iloc[:,-1]
gX = gks.drop(["Player", "Pos","Value"], axis=1)
gY = gks.iloc[:,-1]


# SPlit
trainfX, testfX, trainfY, testfY = train_test_split(fX, fY, test_size=0.25, random_state=38)
trainmX, testmX, trainmY, testmY = train_test_split(mX, mY, test_size=0.25, random_state=38)
traindX, testdX, traindY, testdY = train_test_split(dX, dY, test_size=0.25, random_state=38)
traingX, testgX, traingY, testgY = train_test_split(gX, gY, test_size=0.25, random_state=38)
```

All of the players listed above were "incorrectly" predicted by over 25 (million). The first 8 were predicted by the model that was run over the entire set, while the rest were predicted by the models over the set of their singular position. However, when looking into why these players were predicted "incorrectly", the model was actually looking at the stats correctly and there were clear intanables involved in each players' value. 

As shown by the chart of "Age vs Value", after the age of 30, the value of a player is almost always on a decline. No team wants to buy a player over 30 because of their age, which lowers their value significantly, but it doesn't mean that the player will play poorly. 

Sergio Aguero, David Silva, and Fernandinho are all over the age of 30, but they all were instrumental in helping their team, Manchester City, in winning the Premier league in the season the stats came from. Because of how well they played throughout the season, their stats were alike players who had much higher values than they did. 

On the side of underpredictions, there is Phil Foden and Bukayo Saka, who are two of England's young star players. Because of how young they are, as well as the fact that they are British and playing in the best British League, they are bound to have an extremely high market value. But based on their stats alone, the model predicts that they are still very good players, but are not entirely worth the value they are given. On the other hand, Mykhailo Mudryk is a young Ukranian player who played amazingly in the Ukranian league, but when an English team bought him and he played in the Premier League, he underperformed massively, which is shown by the stats and is why the model predicted him at only 4 million. His value given however, is still relatively accepted because he is a young player that has a lot of talent. But the adjustment for some people is hard, which is another intangable which cannot be grasped by data (He is doing much better now than he did that first season).

There are then players such as Leandro Trossard, Dominic Calvert-Lewin, Manuel Akanji, and John Stones who are players that have values which should be higher. Leasndro Trossard had a great season based on his stats and he was still being undervalued, so the season after that another team saw this and got him for a discounted price and he has performed well for them. Another is Manuel Akanji which was bought from a different league for a decent price and played amazingly for Manchester City in his first season. His value increased after the season, but the model predicted that his value should have increased more.

There are also the players such as Armando Broja and Faustino Anjorin who only played a few minutes in a season so their stats are not representative of their actual value and they are massively overpredicted.

# Conclusion

After applying multiple models to the dataset I achieved a mean absolute error of 8.1 million (4% of the range of values) with the XGBoost regression model. 48 of the 229 predictions were within 1% of the range (1.8 million) while 153 were within 5% (9 million). 

When splitting the dataset into four smaller subsets by position, there was an increse in the error in forwards while the defenders and goalkeeper errors decreased. The error in forwards being so much larger is because they are the most volatile players when it comes to their value. Forwards' jobs are to score goals and win the game for their team which makes their average value so much greater than other positions. BUt if they have one bad season (sometimes even one bad game) their value can drop immensely.

For defenders and goalkeepers they are still under the same scrutiny as the forwards but their average values are much lower and less volatile than forwards which is why the models were predicting players much more closer to their given values. The model still predicted defenders that played more than what their value is worth as well.

Overall, the models do a very good job predicting how much a player's value is while also being able to point out outliers and more accurately value them based on solely their playing statistics.
